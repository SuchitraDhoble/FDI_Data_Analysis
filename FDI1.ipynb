```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path='D:\\Unified_Mentor\\Working_File\\FDI data.csv'
FD=pd.read_csv(file_path)
```

```python
FD.head()
```

```python
FD.info()
```

```python
FD.dropna(inplace=True)
```

```python
for col in FD.columns[1:]:
    FD[col]= pd.to_numeric(FD[col].astype(str).str.replace(',',''),errors='coerce')
```

```python
FD.fillna(0,inplace=True)
print(FD.head())
```

```python
FD.info()
```

```python
descriptive_stats=FD.describe()
print(descriptive_stats)
```

```python
sector_grouped=FD.groupby('Sector').sum()
print(sector_grouped)
```

```python
#set plot style
sns.set(style="dark")
#Distribution of FDI over the years
plt.figure(figsize=(10,7))
for col in FD.columns[1:]:
    sns.histplot(FD[col], kde=True,label=col)
plt.title('Distribution of FDI over the Years')
plt.xlabel('FDI in Crores')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#total FDI by Sector 
total_fdi_by_sector=sector_grouped.sum(axis=1).sort_values(ascending=False)
plt.figure(figsize=(15,45))
sns.barplot(x=total_fdi_by_sector.values,y=total_fdi_by_sector.index)
plt.title('Total FDI by Sector')
plt.xlabel('Total FDI in Crores')
plt.ylabel('Sector')
plt.show()
```

```python
#Yearly FDI trends for top 5 sectors
top_5_sectors=total_fdi_by_sector.head(5).index
plt.figure(figsize=(14,10))
for sector in top_5_sectors:
    plt.plot(FD[FD['Sector']==sector].set_index('Sector').T,label=sector)
plt.title('Yearly FDI Trends for Top 5 Sectors')
plt.xlabel("Year")
plt.ylabel('FDI in Crores')
plt.legend()
plt.show
```

```python
plt.figure(figsize=(14, 10))
for sector in FD['Sector']:
    sector_data = FD[FD['Sector'] == sector].set_index('Sector').T
    plt.plot(sector_data.index, sector_data.values, label=sector)
plt.title('Yearly FDI Trends for All Sectors')
plt.xlabel('Year')
plt.ylabel('FDI in Crores')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sector_year_data = FD.set_index('Sector').T
sns.heatmap(sector_year_data, cmap='YlGnBu', linewidths=.5)
plt.title('Heatmap of FDI Over Years by Sector')
plt.xlabel('Sector')
plt.ylabel('Year')
plt.show()
```

```python
plt.figure(figsize=(5, 25))
melted_df = FD.melt(id_vars=['Sector'], var_name='Year', value_name='FDI in Crores')
sns.boxplot(x='FDI in Crores', y='Sector', data=melted_df, palette='coolwarm')
plt.title('Box Plot of FDI by Sector')
plt.xlabel('FDI in Crores')
plt.ylabel('Sector')
plt.show()
```

```python
top_n = 5
top_n_sectors = total_fdi_by_sector.head(top_n).index
top_n_df = FD[FD['Sector'].isin(top_n_sectors)].set_index('Sector').T

plt.figure(figsize=(14, 8))
top_n_df.plot.area(figsize=(14, 8), colormap='tab20')
plt.title('Top 5 Sectors Over the Years')
plt.xlabel('Year')
plt.ylabel('FDI in Crores')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Define the number of bottom sectors to display
bottom_n = 5
# Total FDI by Sector
total_fdi_by_sector = FD.set_index('Sector').sum(axis=1).sort_values(ascending=False)

# Bottom N sectors by FDI
bottom_n_sectors = total_fdi_by_sector.tail(bottom_n)

plt.figure(figsize=(12, 8))
sns.barplot(x=bottom_n_sectors.values, y=bottom_n_sectors.index, palette='viridis')
plt.title(f'Total FDI by Bottom {bottom_n} Sectors (2000-01 to 2016-17)')
plt.xlabel('Total FDI in Crores')
plt.ylabel('Sector')
plt.show()
```

```python
# Define the number of top sectors to display
top_n = 10

# Top N sectors by FDI
top_n_sectors = total_fdi_by_sector.head(top_n)

plt.figure(figsize=(10, 15))
plt.pie(top_n_sectors, labels=top_n_sectors.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', top_n))
plt.title(f'Top {top_n} Sectors by Total FDI (2000-01 to 2016-17)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
